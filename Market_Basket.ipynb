```python
#Creating a data frame with only fruits and vegetables.

selected_columns_df = raw_df[['Lemons', 'Onions', 'Beets', 'Sweat_Potatoes','Lettuce', 'Green_Pepper', 'Frozen_Broccoli', 'Canned_Peas', 'Whole_Green_Beans', 'Plums', 'Mushrooms', 'Dried_Apples', 'Carrots', 'Raisins', 'Frozen_Cauliflower', 'Tomatoes', 'Canned_Mixed_Fruit', 'Limes', 'Cabbage', 'Dried_Apricots', 'Peaches', 'Apples', 'Frozen_Carrots', 'Bananas', 'Pepper', 'Oranges', 'Cauliflower', 'Black-eyed_Peas', 'Mandarin_Oranges', 'Frozen_Peas', 'Garlic', 'Dried_Fruit_Mix', 'Broccoli', 'Mixed_Vegetables', 'Red_Pepper', 'Canned_Beets', 'Peas', 'Canned_Peaches', 'Sweet_Potatoes', 'Canned_Tomatoes', 'Green_Beans', 'Potatoes']] 
```

```python
# import the apriori algorithm
from mlxtend.frequent_patterns import apriori
```

#### Identify frequent itemsets, i.e., items or item sets popular in the data sets.

##### Minimum support required in the data sets

```python
# If a very high value of support is set, there won't be any combinations.

minimum_required_support=0.015
```

```python
frequent_itemsets = apriori(selected_columns_df,min_support=minimum_required_support,use_colnames=True)
```

```python
# The following items have support values greater than the minimum support models.

frequent_itemsets
```

```python
max_support_value = frequent_itemsets['support'].max()
max_support_value
```

```python
itemsets_with_max_support_value = frequent_itemsets[frequent_itemsets['support'] == max_support_value]['itemsets'].tolist()

print("Maximum Support:", max_support_value)
print("Itemsets with Maximum Support:", itemsets_with_max_support_value)
```

```python
#Apply association mining algorithm using this support
frequent_itemsets= apriori(selected_columns_df, min_support=minimum_required_support, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
## Set rules from the frequent itemsets that match the desired lift.

## Set the required lift

minimum_required_lift = 1.1
rules = association_rules(frequent_itemsets,metric="lift", min_threshold=minimum_required_lift)
```

```python
rules
```

```python
print("Total number of rules generated by the algorithm with given support and lift:",len(rules))
```

```python
# Set the minimum threshold for confidence
minimum_required_confidence = 0.5
```

```python
# Filter rules based on confidence and lift threshold
filtered_rules = rules[(rules['confidence'] >= minimum_required_confidence)]
```

```python
filtered_rules
```

```python
print("Total number of filtered rules to meet the lift, confidence and support criteria:",len(filtered_rules))
```

```python
# Sorting the rules based on lift, confidence, and support (all in descending order)

sorted_rules = filtered_rules.sort_values(by=['lift','confidence','support'],ascending=False)
```

```python
sorted_rules
```

```python
# Selecting the top 5 rules

top_5_rules = sorted_rules.head(5)
```

```python
# Print only the support, confidence, and lift column

top_5_rules[['antecedents','consequents', 'support', 'confidence', 'lift']]
```

### Inference:

Lift of frequently bought item no 21: Canned_Peaches is greater than the item no 103. 
People who buy canned-peaches also buy onions.
Simialrly,  customers who buy bananas and potatoes in this order also buy onions.

### Business decision:

The grocery store can adjust its layout so that onions are kept next to or close to where canned_peaches are stored. Similarly. it can keep onions, next to bananas and potatoes.
